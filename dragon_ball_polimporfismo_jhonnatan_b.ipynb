```python
# Programa en Python que simula personajes de Dragon Ball con clases y polimorfismo

class Personaje:
    def __init__(self, nombre, raza, genero):
        self.nombre = nombre
        self.raza = raza
        self.genero = genero  # Puede ser "Masculino" o "Femenino"

    # Método que debe ser implementado en las subclases
    def transformar(self, condicion):
        pass

    # Método para "procrear" un hijo
    def procrear(self, otro_personaje):
        pass

class Saiyajin(Personaje):
    def __init__(self, nombre, tiene_cola=True, genero="Masculino"):
        super().__init__(nombre, "Saiyajin", genero)
        self.tiene_cola = tiene_cola

    def transformar(self, condicion):
        if condicion.lower() == "luna llena" and self.tiene_cola:
            print(f"{self.nombre} se transforma en un mono gigante porque tiene cola y ve la luna llena.")
        else:
            print(f"{self.nombre} no se transforma.")

    def quitar_cola(self):
        self.tiene_cola = False
        print(f"{self.nombre} ha perdido su cola y ya no puede transformarse en mono gigante.")

    def procrear(self, otro_personaje):
        # Solo los hombres y mujeres pueden procrear entre sí
        if self.genero == "Masculino" and otro_personaje.genero == "Femenino":
            # Si el otro personaje es un Humano, el hijo será híbrido
            if isinstance(otro_personaje, Humano):
                hijo_nombre = self.nombre + " Jr."
                return Hibrido(hijo_nombre)
            # Si el otro personaje también es un Saiyajin, el hijo será Saiyajin
            elif isinstance(otro_personaje, Saiyajin):
                hijo_nombre = "Goten"  # Ejemplo de hijo de Saiyajin
                return Saiyajin(hijo_nombre)
        elif self.genero == "Femenino" and otro_personaje.genero == "Masculino":
            # Si el otro personaje es un Humano, el hijo será híbrido
            if isinstance(otro_personaje, Humano):
                hijo_nombre = "Pan"  # Ejemplo de hija híbrida
                return Hibrido(hijo_nombre)
            # Si el otro personaje es un Saiyajin, el hijo será Saiyajin
            elif isinstance(otro_personaje, Saiyajin):
                hijo_nombre = "Goten"  # Ejemplo de hija Saiyajin
                return Saiyajin(hijo_nombre)
        return None

class Humano(Personaje):
    def __init__(self, nombre, genero="Masculino"):
        super().__init__(nombre, "Humano", genero)

    def transformar(self, condicion):
        print(f"{self.nombre} es un humano y no puede transformarse.")

    def procrear(self, otro_personaje):
        # Solo los hombres y mujeres pueden procrear entre sí
        if self.genero == "Masculino" and otro_personaje.genero == "Femenino":
            # Si el otro personaje es un Saiyajin, el hijo será híbrido
            if isinstance(otro_personaje, Saiyajin):
                hijo_nombre = "Pan"  # Ejemplo de hijo híbrido
                return Hibrido(hijo_nombre)
            # Si el otro personaje también es Humano, el hijo será Humano
            elif isinstance(otro_personaje, Humano):
                hijo_nombre = "Videl"  # Ejemplo de hijo humano
                return Humano(hijo_nombre)
        elif self.genero == "Femenino" and otro_personaje.genero == "Masculino":
            # Si el otro personaje es un Saiyajin, el hijo será híbrido
            if isinstance(otro_personaje, Saiyajin):
                hijo_nombre = "Pan"  # Ejemplo de hija híbrida
                return Hibrido(hijo_nombre)
            # Si el otro personaje es un Humano, el hijo será Humano
            elif isinstance(otro_personaje, Humano):
                hijo_nombre = "Videl"  # Ejemplo de hija humana
                return Humano(hijo_nombre)
        return None

class Hibrido(Personaje):
    def __init__(self, nombre, genero="Masculino"):
        super().__init__(nombre, "Híbrido Saiyajin-Humano", genero)

    def transformar(self, condicion):
        if condicion.lower() == "luna llena":
            print(f"{self.nombre} tiene una transformación parcial debido a sus genes Saiyajin e intenta transformarse.")
        else:
            print(f"{self.nombre} no se transforma.")

    def procrear(self, otro_personaje):
        # Solo los hombres y mujeres pueden procrear entre sí
        if self.genero == "Masculino" and otro_personaje.genero == "Femenino":
            # Los híbridos pueden tener hijos híbridos o un hijo humano, dependiendo del cruce
            if isinstance(otro_personaje, Humano):
                hijo_nombre = "Pan"  # Ejemplo de hijo híbrido
                return Hibrido(hijo_nombre)
            elif isinstance(otro_personaje, Saiyajin):
                hijo_nombre = "Goten"  # Ejemplo de hijo Saiyajin
                return Saiyajin(hijo_nombre)
        elif self.genero == "Femenino" and otro_personaje.genero == "Masculino":
            # Los híbridos pueden tener hijos híbridos o un hijo humano, dependiendo del cruce
            if isinstance(otro_personaje, Humano):
                hijo_nombre = "Pan"  # Ejemplo de hija híbrida
                return Hibrido(hijo_nombre)
            elif isinstance(otro_personaje, Saiyajin):
                hijo_nombre = "Goten"  # Ejemplo de hija Saiyajin
                return Saiyajin(hijo_nombre)
        return None

class Namekusein(Personaje):
    def __init__(self, nombre, genero="Masculino"):
        super().__init__(nombre, "Namekusein", genero)

    def transformar(self, condicion):
        print(f"{self.nombre} es un Namekusein y no puede transformarse como un Saiyajin.")

    def procrear(self, otro_personaje):
        # Solo los hombres y mujeres pueden procrear entre sí
        if self.genero == "Masculino" and otro_personaje.genero == "Femenino":
            # Si el otro personaje es un Humano o Saiyajin, el hijo será un Namekusein
            hijo_nombre = self.nombre + " Jr."  # Ejemplo de hijo de Namekusein
            return Namekusein(hijo_nombre)
        elif self.genero == "Femenino" and otro_personaje.genero == "Masculino":
            hijo_nombre = self.nombre + " Jr."  # Ejemplo de hija Namekusein
            return Namekusein(hijo_nombre)
        return None

# Crear los personajes
goku = Saiyajin("Goku", tiene_cola=True, genero="Masculino")
gohan = Hibrido("Gohan", genero="Masculino")
krillin = Humano("Krillin", genero="Masculino")
vegeta = Saiyajin("Vegeta", tiene_cola=True, genero="Masculino")
piccolo = Namekusein("Piccolo", genero="Masculino")  # Ahora Piccolo es un Namekusein
videl = Humano("Videl", genero="Femenino")
chi_chi = Humano("Chi-Chi", genero="Femenino")

# Función para seleccionar personajes
def seleccionar_personaje():
    print("Selecciona un personaje para interactuar:")
    print("1. Goku (Saiyajin)")
    print("2. Gohan (Híbrido Saiyajin-Humano)")
    print("3. Krillin (Humano)")
    print("4. Vegeta (Saiyajin)")
    print("5. Piccolo (Namekusein)")
    print("6. Videl (Humano Femenino)")
    print("7. Chi-Chi (Humano Femenino)")
    seleccion = input("Ingresa el número del personaje: ")

    if seleccion == "1":
        return goku
    elif seleccion == "2":
        return gohan
    elif seleccion == "3":
        return krillin
    elif seleccion == "4":
        return vegeta
    elif seleccion == "5":
        return piccolo
    elif seleccion == "6":
        return videl
    elif seleccion == "7":
        return chi_chi
    else:
        print("Selección inválida. Escoge un número del 1 al 7.")
        return seleccionar_personaje()

# Función para crear un hijo entre dos personajes
def crear_hijo():
    print("\nSelecciona el primer personaje:")
    padre = seleccionar_personaje()

    print("\nSelecciona el segundo personaje:")
    madre = seleccionar_personaje()

    # Los personajes pueden tener un hijo si son hombre y mujer
    hijo = padre.procrear(madre)
    if hijo:
        print(f"\nEl hijo de {padre.nombre} y {madre.nombre} se llama {hijo.nombre} y es un {hijo.raza}.")
    else:
        print("\nNo se puede crear un hijo entre estos personajes (deben ser hombre y mujer).")

# Ejecutar la creación de un hijo
crear_hijo()
```

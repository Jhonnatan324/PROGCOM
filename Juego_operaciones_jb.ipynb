```python
import random

# Función para generar preguntas
def generar_pregunta(operacion):
    num1 = random.randint(1, 10)
    num2 = random.randint(1, 10)

    if operacion == "suma":
        respuesta = num1 + num2
        pregunta = f"{num1} + {num2} = "
    elif operacion == "resta":
        respuesta = num1 - num2
        pregunta = f"{num1} - {num2} = "
    elif operacion == "multiplicacion":
        respuesta = num1 * num2
        pregunta = f"{num1} * {num2} = "
    elif operacion == "division":
        # Evitamos la división por cero
        while num2 == 0:
            num2 = random.randint(1, 10)
        respuesta = round(num1 / num2, 2)  # Redondeamos a 2 decimales
        pregunta = f"{num1} / {num2} = "
    return pregunta, respuesta

# Función principal del juego
def juego_matematicas():
    print("¡Bienvenido al juego de matemáticas!\n")
    print("Elige el tipo de operación:")
    print("1. Suma")
    print("2. Resta")
    print("3. Multiplicación")
    print("4. División")
    print("5. Mezcladas\n")

    opcion = input("Ingresa el número de tu elección: ")
    opciones = {"1": "suma", "2": "resta", "3": "multiplicacion", "4": "division", "5": "mezcladas"}

    if opcion not in opciones:
        print("Opción no válida. Intenta de nuevo.")
        return

    correctas = 0
    while correctas < 10:
        operacion = opciones[opcion]

        # Si elige mezcladas, selecciona una operación al azar
        if operacion == "mezcladas":
            operacion = random.choice(["suma", "resta", "multiplicacion", "division"])

        pregunta, respuesta = generar_pregunta(operacion)

        try:
            respuesta_usuario = float(input(pregunta))
            if respuesta_usuario == round(respuesta, 2):
                print("¡Correcto!\n")
                correctas += 1
            else:
                print(f"Incorrecto. La respuesta correcta era {respuesta}\n")
                correctas = 0
        except ValueError:
            print("Por favor, ingresa un número válido.")

    print("¡Felicidades! Respondiste correctamente a 10 preguntas seguidas. ¡Has ganado!")

# Ejecutar el juego
juego_matematicas()
```
